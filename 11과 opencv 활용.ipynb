# 2.이미지 편집하기

## (1) 사진 크기 조정

* cv2.resize(img, (width, height))
* width, height는 자연수여야 함.

```python
height, width, channel = img.shape

img2 = cv2.resize(img, (215, 250))
plt.imshow(img2)
```

## (2) 대칭

* 대칭 축은 상수를 입력해 대칭할 축을 설정할 수 있습니다.
* flipCode < 0은 XY 축 대칭(상하좌우 대칭)을 적용합니다.
* flipCode = 0은 X 축 대칭(상하 대칭)을 적용합니다.
* flipCode > 0은 Y 축 대칭(좌우 대칭)을 적용합니다.

```python
print('original')
plt.imshow(img)

img2 = cv2.flip(img, 0)
print('상하대칭')
plt.imshow(img2)

img2 = cv2.flip(img, -1)
print('상하좌우대칭')
plt.imshow(img2)

img2 = cv2.flip(img, 1)
print('좌우대칭')
plt.imshow(img2)
```

## (4) 흐리게(Blur)
* 흐림 효과는 
    * 각 픽셀에 대해 커널을 적용해서,
    * 커널 범위 내의 픽셀 평균값으로 바꾸게 됩니다.
* cv2.blur(이미지, (커널 가로 크기, 커널 세로 크기))
    * 커널 크기가 클 수록 흐림 효과가 더 커지게 됩니다.

```python
img2 = cv2.blur(img, (10, 10))
cv2_imshow(img2)
```

## (5) 이미지 Blending
* 두 이미지를 겹쳐서 하나의 이미지로 만들 수 있습니다.
    * cv2.addWeighted(이미지1, 이미지1투명도, 이미지2, 이미지2투명도, 0)

```python
img1 = cv2.imread('man.png')
img2 = cv2.imread('lion.png')
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))

for a in np.linspace(0, 1, 11) :
    dst = cv2.addWeighted(img1, a, img2, (1-a), 0)
    print('img1',a, 'img2', 1-a)
    cv2_imshow(dst)
```

```python
dst = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
cv2_imshow(dst)
```
